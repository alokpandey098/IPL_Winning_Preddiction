```python
delivery_dataset['total_runs_y'] = delivery_dataset['total_runs_y'].astype(int)
delivery_dataset['current_score'] = delivery_dataset.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_dataset.head()
```

```python
delivery_dataset['Runs left'] = delivery_dataset['total_runs_x'] - delivery_dataset['current_score']+1
```

```python
delivery_dataset['Balls left'] = 126 - (delivery_dataset['over']*6 + delivery_dataset['ball'])
```

```python
delivery_dataset.head()
```

```python
delivery_dataset['player_dismissed'] = delivery_dataset['player_dismissed'].fillna("0")
delivery_dataset['player_dismissed'] = delivery_dataset['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_dataset['player_dismissed'] = delivery_dataset['player_dismissed'].astype('int')
wickets = delivery_dataset.groupby('match_id')['player_dismissed'].cumsum().values
delivery_dataset['wickets'] = 10 - wickets
delivery_dataset.head()
```

```python
delivery_dataset.head()
```

```python
delivery_dataset['CRR'] = (delivery_dataset['current_score']*6)/(120 - delivery_dataset['Balls left'])
```

```python
delivery_dataset['RRR'] = (delivery_dataset['Runs left']*6)/delivery_dataset['Balls left']
```

```python
delivery_dataset.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_dataset['result'] = delivery_dataset.apply(result,axis=1)
```

```python
final_dataset = delivery_dataset[['batting_team','bowling_team','city','Runs left','Balls left','wickets','total_runs_x','CRR','RRR','result']]
```

```python
final_dataset.head()
```

```python
final_dataset = final_dataset.sample(final_dataset.shape[0])
final_dataset.sample()
```

```python
final_dataset.dropna(inplace=True)
```

```python
final_dataset = final_dataset[final_dataset['Balls left'] != 0]
```

```python
X = final_dataset.iloc[:,:-1]
y = final_dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
final_dataset.info()
```

```python
final_dataset.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

# Grid searchcv

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.best_params_
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='newton-cg',C=100))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('model.pkl','wb'))
```
